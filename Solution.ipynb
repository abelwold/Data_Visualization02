```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def problem2(file):
    # Read data
    data = pd.read_csv(file)
    # Make the plot background white
    sns.set_style(style='white')
    fig, ax1 = plt.subplots(figsize=(6, 7))
    # Create a twin Axes sharing the yaxis 
    ax2 = ax1.twinx()
    # First data plot
    lns1 = ax1.plot(data.rest, 'r-s',lw=1,ms=3.5,label="rest (left-axis)")
    # Second data plot
    lns2 = ax2.plot(data.china, 'b-v',lw=1,ms=3.5,label="china (right-axis)")
    ax1.set_xlabel('Days')
    # Assign both data's plot to variable
    lns = lns2+lns1
    # Get lables of both data's
    labs = [l.get_label() for l in lns]
    # Assign the legends 
    ax1.legend(lns, labs, loc=0, frameon=False,fontsize=13.5)
    # Remove spines
    ax1.spines['top'].set_visible(False)
    sns.despine(ax=ax2)
    ax1.set_title('Number of cases')
    plt.tight_layout()
    # Save plot as png with a file name 'Number_of_cases'
    plt.savefig('Number_of_cases.png')
    plt.show()
```

```python
problem2("corona_china_vs_rest.csv")
```
